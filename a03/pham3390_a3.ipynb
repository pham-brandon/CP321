```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
# Task 1 A
def data_download(ticker, date, interval):
    start_date = pd.to_datetime(date)
    end_date = start_date + pd.Timedelta(days=30)
    data = yf.download(ticker, start=start_date, end=end_date, interval=interval)
    return data
```

```python
# Task 1 B
# Test Case 1
ticker1 = "AAPL"
date1 = "2025-02-14"
interval1 = "5m"
data1 = data_download(ticker1, date1, interval1)
display(data1.head())

# Test Case 2
ticker2 = "GOOG"
date2 = "2025-02-14"
interval2 = "5m"
data2 = data_download(ticker2, date2, interval2)
display(data2.head())
```

```python
# Task 2 A
def volume_analysis(data, ticker, date, interval):
    plt.figure(figsize=(12, 6))
    plt.bar(data.index, data[('Volume', ticker)], color='cornflowerblue')

    plt.xlabel('Time')
    plt.ylabel('Volume')
    plt.legend([f'{ticker} Volume'])
    plt.tight_layout()
    plt.title(f'Volume Analysis of {ticker} on {date} using {interval} interval')
    plt.show()
```

```python
# Task 2 B
# Test Case 1
ticker1 = "AAPL"
date1 = "2025-01-01"
interval1 = "5m"
data1 = data_download(ticker1, date1, interval1)
volume_analysis(data1, ticker1, date1, interval1)

# Test Case 2
ticker2 = "GOOG"
date2 = "2025-01-01"
interval2 = "5m"
data2 = data_download(ticker2, date2, interval2)
volume_analysis(data2, ticker2, date2, interval2)
```

```python
# Task 3 A
def price_analysis(data, ticker, date, interval):
    plt.figure(figsize=(12, 6))
    data['Mean_Price'] = data[['Close', 'High', 'Low', 'Open']].mean(axis=1)
    plt.plot(data.index, data['Close'], label='Close Price', color='darkorange', linestyle='--')
    plt.plot(data.index, data['Mean_Price'], label='Mean Price', color='blue')

    plt.xlabel('Time')
    plt.ylabel('Price') 
    plt.legend()
    plt.grid(True, linestyle='-', alpha=0.7)
    plt.tight_layout()
    plt.title(f'Price Analysis of {ticker} on {date} using {interval} interval')
    plt.show()
```

```python
# Task 3 B
# Test Case 1
ticker1 = "AAPL"
date1 = "2025-01-01"
interval1 = "5m"
data1 = data_download(ticker1, date1, interval1)
price_analysis(data1, ticker1, date1, interval1)

# Test Case 2
ticker2 = "GOOG"
date2 = "2025-01-01"
interval2 = "5m"
data2 = data_download(ticker2, date2, interval2)
price_analysis(data2, ticker2, date2, interval2)
```

```python
# Task 4 A
def violin_plots(data, ticker, date, interval):
    plt.figure(figsize=(12, 6))
    columns = ['Close', 'Open', 'Low', 'High']
    data_plt = [data[column].dropna().values.flatten().tolist() for column in columns]
    plt.violinplot(data_plt, showmeans=True, showmedians=True, showextrema=True)
    
    plt.xlabel('Price Types')
    plt.ylabel('Price')
    plt.xticks(np.arange(1, len(columns) + 1), columns)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.title(f'Violin Plot of {ticker} on {date} using {interval} interval')
    plt.show()
```

```python
# Task 4 B
# Test Case 1
ticker1 = "AAPL"
date1 = "2025-01-01"
interval1 = "5m"
data1 = data_download(ticker1, date1, interval1)
violin_plots(data1, ticker1, date1, interval1)

# Test Case 2
ticker2 = "GOOG"
date2 = "2025-01-01"
interval2 = "5m"
data2 = data_download(ticker2, date2, interval2)
violin_plots(data2, ticker2, date2, interval2)
```
